```python
# 라이브러리 사용

import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거의 데이터 준비

파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()
```

```python
# 원핫인코딩

인코딩 = pd.get_dummies(아이리스)
인코딩.head()

# 종속변수가 범주형(양적 X)
# pd.get_dummies() -> 자료들 중 범주형인 것들만 골라 1, 0으로 바꿔주는 코드
```

```python
# 칼럼이름 출력
print(인코딩.columns)
```

```python
# 독립변수, 종속변수

독립 = 인코딩[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
종속 = 인코딩[['품종_setosa', '품종_versicolor',
       '품종_virginica']]
print(독립.shape, 종속.shape)
```

```python
# 2.모델의 구조 만들기

X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation="swish")(X)
H = tf.keras.layers.Dense(8, activation="swish")(H)
H = tf.keras.layers.Dense(8, activation="swish")(H)
Y = tf.keras.layers.Dense(3, activation = 'softmax')(H)
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')

# 분류모델 - sigmoid, softmax 차이 알아보기
# y의 최대값 1, 최소값 0 -> y=softmax(w1x1+w2x2+..+b)로 감싸주면 됨
# activation: 활성화 함수
# 분류에 사용하는 loss - crossentropy, 회귀에 사용하는 loss - mse
# metrics='accuracy' -> 정확도도 함께 볼 수 있음
```

```python
# 모델 구조 확인
model.summary()
```

```python
# 3. 데이터로 모델 학습(FIT)시키기

model.fit(독립, 종속, epochs=100)
```

```python
# 4. 모델 이용하기

#맨 처음 데이터 5개
print(model.predict(독립[:5]))
print(종속[:5])

#맨 마지막 데이터 5개
print(model.predict(독립[-5:]))
print(종속[-5:])
```
